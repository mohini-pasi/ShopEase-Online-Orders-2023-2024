
# Task 4 — Short Written Summary

• The single most important insight from the data
Electronics is the top revenue category.

• One year-on-year observation (2023 vs 2024)
2023 vs 2024: Revenue grew in 2024, especially during holiday months.

• One underperforming area — region, category, or product — and a possible reason
Underperforming area: The East and West regions have the highest number of cancelled orders (35 each). This could be due to payment issues or customers changing their minds.

Possible reason: Customers in these regions might be using payment methods that fail (like COD or Net Banking) or facing longer delivery times.

• One actionable recommendation for the business
Actionable recommendation: Check why cancellations are high in East and West. Send payment confirmation messages and offer faster delivery options to reduce cancellations.
