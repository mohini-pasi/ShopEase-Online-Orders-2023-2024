here the order_date data type is object so i am 1st converting it into datetime

## 4. What is the average delivery_days for Delivered orders, broken down by region?

```python
delivered_order = df[df['order_status'] == 'Delivered']
```

```python
avg_delivery = delivered_order.groupby('region')['delivery_days'].mean()
print(avg_delivery)
```

## 5. Which age group (e.g. 18–25, 26–35, 36–45, 46+) placed the most orders?

```python
bins = [18, 25, 35, 45, 60]
labels = ['18-25', '26-35', '36-45', '46+']
```

```python
df['age_group'] = pd.cut(df['customer_age'], bins=bins, labels=labels)
age_group_counts = df['age_group'].value_counts()
print(age_group_counts)
print("Total placed order by age group :",age_group_counts.sum())
print("Total placed order by missing age group :",df['age_group'].isnull().sum())
```
